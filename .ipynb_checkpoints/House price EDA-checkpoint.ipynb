```python
attrs = corrMatrix.iloc[:-1, :-1] # all except last one, which is SalePrice

threshold = 0.5
important_corrs = (attrs[abs(attrs) > threshold][attrs != 1.0]).unstack().dropna().to_dict()
# attrs[abs(attrs) > threshold]: return all number > threshold
# [attrs != 1.0]: return all number != 1.0
# unstack(): return all numbers in a series
# important_corrs is a dict which has double key values 

unique_important_corrs = pd.DataFrame(
    list(set([(tuple(sorted(key)), important_corrs[key]) for key in important_corrs])), 
    columns = ['Attribute Pair', 'Correlation'])
# use [tuple() for x in range(n)] in list comprehension to generate multiple item.
# [(tuple(sorted(key)), important_corrs[key]) for key in important_corrs] return a list with a sorted key and corr
# set only save unique value.


# sorted by absolute value
unique_important_corrs = unique_important_corrs.loc[
    abs(unique_important_corrs['Correlation']).argsort()[::-1]]

unique_important_corrs
```

# Missing Data

    the way I try to deal with missing data is not to delete it but to fill it with proper indication. if a column relates to others, for e.g., 'PoolQC' only has data when 'PoolArea' > 0. So I'll fill it with 'No' for 'PoolArea' == 0. 
    however, for column has no connection to other and not important for 'SalePrice', I may delete it.

```python
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

### 'LotFrontage'
        from the plot and the p-value from the t-test, we can deduce that we should fill the NaN with mode or mean and not with 'NA' or 'No'. In this case, I combine all the train and test, and fill the NaNs with the mode.

```python
# it's either NaN or > 0
cond= train['LotFrontage'].isnull()
print train.groupby(cond).describe()['SalePrice']

# to do
# plot the SalePrice distribution (depends on 'LotFrontage')
# run statistics test

LotFrontageH= train['SalePrice'][train['LotFrontage'] > 0]
LotFrontageN= train['SalePrice'][train['LotFrontage'].isnull()]

plt.subplots(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.distplot(LotFrontageH)
plt.plot([np.mean(LotFrontageH), np.mean(LotFrontageH)], [0, 0.000008], linewidth=2, color = 'r')
plt.subplot(1, 2, 2)
sns.distplot(LotFrontageN)
plt.plot([np.mean(LotFrontageN), np.mean(LotFrontageN)], [0, 0.000008], linewidth=2, color = 'r')
plt.show()

print stats.ttest_ind(LotFrontageH, LotFrontageN)

temp= pd.concat([train['LotFrontage'], test['LotFrontage']])
modeFill= stats.mode(temp)[0][0]

train['LotFrontage'].replace(np.nan, modeFill, inplace= True)
test['LotFrontage'].replace(np.nan, modeFill, inplace= True)
```

### Garage
